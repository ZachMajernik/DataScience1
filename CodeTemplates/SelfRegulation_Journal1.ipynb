```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
tables = []
soups = []
```

```python
# URL of the Wikipedia article you want to scrape
url = "https://en.wikipedia.org/wiki/The_world%27s_100_most_threatened_species"

# Make an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup0 = BeautifulSoup(response.content, 'html.parser')
soups.append(soup0)

table = soups[0].find_all('table')

tables.append(table)
```

```python
# URL of the Wikipedia article you want to scrape
url = "https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States"

# Make an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup1 = BeautifulSoup(response.content, 'html.parser')
soups.append(soup1)

table = soups[1].find_all('table')

tables.append(table)
```

```python
tables[0]
```

```python
df = pd.DataFrame()

tableHeads = soups[0].find_all('th')
rowCount = len(soups[0].find_all('td'))/len(tableHeads)
for i in range(len(tableHeads)):
    var = str(tableHeads[i])
    var = var.split('>')[1].split('<')[0].split('\n')[0]
    tableHeads[i] = var
    df[var] = var
```
